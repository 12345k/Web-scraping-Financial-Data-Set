```python
import pandas as pd
import urllib.request as urllib2
from bs4 import BeautifulSoup
import requests
```

```python
page=urllib2.urlopen("http://www.moneycontrol.com/mutual-funds/performance-tracker/ranks/diversified-equity-2.html")
soup_1 = BeautifulSoup(page)
Name=[]
Links=[]
all_links=soup_1.find_all('a',class_='bl_12')
for i in all_links:
    Links.append(i['href'])
    Name.append(i.find(text=True))
link_list = pd.DataFrame(
    {'Name': Name,
     'Links': Links
     
    })
link_list
```

```python

if "Tata Dividend Yield Fund - Direct Plan (G)" in link_list['Name']:
    print("Tata Dividend Yield Fund - Direct Plan (G)")



link_list = link_list[190:] 
link_list
```

```python
Holding=[]
# web=['/mutual-funds/nav/reliance-growth-fund-rp/MRC008','/mutual-funds/nav/tata-equity-p-e-fund/MTA125']
all_columns_names=['Fund Class','Name', 'Sector' ,'Qty', '%' ,'Funds_Name' ,'Type','Rating']
l=[]
fail=[]
for index_link in link_list['Links']:  
    url="http://www.moneycontrol.com"+index_link #Open the main page of the funds 
    try:
        page=urllib2.urlopen(url) # get the request from the page
        soup_page = BeautifulSoup(page,"lxml")
        a_tags=soup_page.find_all('a') # Read all the a href tags and make list of it
    #     print(a_tags)
        sub_df=[]
        for href_link in a_tags:
    #         print(href_link.find_all(text=True))
            if(href_link.find(text=True) == "Holdings"): # to get the holding link
                holding_url="http://www.moneycontrol.com"+href_link['href']
                try:
                    df=pd.read_html(holding_url,attrs = {'class': 'tblporhd'})  # read html and find the table class name  
                    for i in range(len(df)): # some page has 2 or 4 table
                        s=df[i]#save to seprate dataframe
                        column_name=[]
                        for j in range(len(s.axes[1])):
                            column_name.append(s[j][0])
                        df[i].columns = [column_name]
                        df[i]=df[i].drop(df[i].index[0])
                        s=df[i]
                        df[i]["Type"]=s.columns.values[0]
                        df[i]["Funds_Name"]=soup_page.h1.find(text=True)
                        df[i].rename(columns={df[i].columns.values[0]:'Name'}, inplace=True)
                        df[i].rename(columns={'Value (Rs cr)':'Value'}, inplace=True)
                        check_columns = df[i].columns.values
                        for j in all_columns_names:
                            if j not in check_columns:
                                df[i][j]=""
                        l.append(df[i])
                        sub_df.append(df[i])
                        #print(soup_page.h1.find(text=True))
                except (ValueError, ConnectionResetError) as e:
                    print('This url failed: ' + holding_url)
                    fail.append(holding_url)
    except ConnectionResetError:
        print('This url failed: ' + holding_url)
        fail.append(holding_url)
#         df_seprate= pd.concat(sub_df).reset_index(drop=True)
#         string=soup_page.h1.find(text=True)+".csv"
#         df_seprate.to_csv(string)
                
#                 print(df[i])
            
# main_df= pd.concat(l).reset_index(drop=True)
# main_dfr
```

```python
main_df= pd.concat(l).reset_index(drop=True)
main_df.rename(columns={'Name':'Asset_Name','Type':'Asset_Type',"%":"Percentage"}, inplace=True)
main_df = main_df[['Asset_Name','Asset_Type','Sector','Funds_Name','Rating','Fund Class','Value','Percentage']]
main_df.to_csv("mini_6.csv")
main_df
```
